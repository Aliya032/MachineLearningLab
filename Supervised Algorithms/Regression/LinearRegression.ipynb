# Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x,y)
plt.xlabel('Years after purchase')
plt.ylabel('Rsale value')
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
def myfunc(x):
    return slope*x+intercept
```

```python
mymodel = list(map(myfunc,x))
```

```python
print(slope)
```

```python
print(intercept)
```

```python

print(mymodel)
```

```python
print(std_err)
```

```python
print(r)
```

```python
plt.scatter(x,y)
plt.xlabel('Years after purchase')
plt.ylabel('Resale value')
plt.plot(x, mymodel, lw=4,c='orange')
plt.show()
```
